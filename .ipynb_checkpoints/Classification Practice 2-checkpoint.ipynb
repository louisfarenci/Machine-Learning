This practice notebook makes use of the sckit-learn method, make_classification, to generate data sets with two features.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X, Y = make_classification(n_features = 2, n_redundant = 0, n_informative = 2, n_clusters_per_class=1,hypercube = True)
```

```python
pos = Y==1
neg = Y==0

fig, ax = plt.subplots(1,1)
ax.scatter(X[neg,0],X[neg,1],marker = 'o',label = 'y=0')
ax.scatter(X[pos,0],X[pos,1],marker = 'x', label = 'y=1')
ax.set_xlabel("$x_0$", fontsize = 15)
ax.set_ylabel("$x_1$", fontsize = 15)
fig.suptitle("Traning Set with two informative features")
ax.legend()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
```

```python
lr_model.fit(X,Y)
```

```python
print(f"Learned weights: {lr_model.coef_}")
print(f"Learned bias: {lr_model.intercept_}")
```

```python
w0,w1 = lr_model.coef_[0,0], lr_model.coef_[0,1]
b = lr_model.intercept_
```

```python
bound = lambda x: -(w0*x+b)/(w1)
boundx = np.arange(np.min(X[:,0]),np.max(X[:,0]),0.25)
boundy = bound(boundx)
```

```python
pos = Y==1
neg = Y==0

fig, ax = plt.subplots(1,1)
ax.scatter(X[neg,0],X[neg,1],marker = 'o',label = 'y=0')
ax.scatter(X[pos,0],X[pos,1],marker = 'x', label = 'y=1')
ax.scatter(boundx,boundy,marker=',',s=2, label = 'Decision boundary')
ax.set_xlabel("$x_0$", fontsize = 15)
ax.set_ylabel("$x_1$", fontsize = 15)
fig.suptitle("Traning Set with two informative features")
ax.legend()
```
